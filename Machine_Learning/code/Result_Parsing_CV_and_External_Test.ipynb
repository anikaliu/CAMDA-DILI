```python
models = []
RF_cv = {}
RF_test = {}
SVM_cv = {}
SVM_test = {}
i=0
for file in glob.glob('./**/cv_scores*'):
    print(file)
    model = file.split('/')[1]
    CV_data = pd.read_csv(file)
    #print(CV_data)
    RF_models_cv = CV_data[CV_data['splits'].str.contains("MCNC.RF.1" + "\.")]
    SVM_models_cv = CV_data[CV_data['splits'].str.contains("MCNC.SVM.1" + "\.")]
    print(SVM_models_cv)
    print(SVM_models_cv['balanced_accuracy'].mean())
    

    RF_test['RF_'+model+'_internal_cv'] = [str(round(RF_models_cv.mean(axis=0)[i],3)) +"\u00B1"+ str(round(RF_models_cv.std(axis=0)[i],3)) for i in range(len(RF_models_cv.mean(axis=0)))]
    SVM_test['SVM_'+model+'_internal_cv'] = [str(round(SVM_models_cv.mean(axis=0)[i],3)) +"\u00B1"+ str(round(SVM_models_cv.std(axis=0)[i],3)) for i in range(len(SVM_models_cv.mean(axis=0)))]
    
    
i=0
for file in glob.glob('./**/test_scores*'):
    model = file.split('/')[1]
    print(model)
    Test_data = pd.read_csv(file)
    


    
    
    RF_models_ts = Test_data[Test_data['splits'].str.contains("MCNC.RF.1" + "\.")]
    SVM_models_ts = Test_data[Test_data['splits'].str.contains("MCNC.SVM.1" + "\.")]
    #print(SVM_models_ts)
    #print(SVM_models_ts['balanced_accuracy'].mean())
    
    RF_test['RF_'+model+'_external_test'] = [str(round(RF_models_ts.mean(axis=0)[i],3)) +"\u00B1"+ str(round(RF_models_ts.std(axis=0)[i],3)) for i in range(len(RF_models_ts.mean(axis=0)))] 
    SVM_test['SVM_'+model+'_external_test'] = [str(round(SVM_models_ts.mean(axis=0)[i],3)) +"\u00B1"+ str(round(SVM_models_ts.std(axis=0)[i],3)) for i in range(len(SVM_models_ts.mean(axis=0)))]
    
```
