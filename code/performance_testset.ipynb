```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
#ts scores ECFP
df = pd.read_csv('../data/data_paper/ECFP/ts_scores_FP_yscr.csv', delimiter=',')
df.head()
```

```python
df.sort_values(by=['ROC_AUC'], ascending=False)
```

```python
#plot ROC_AUC for MCNC-RF, MCNC-SVM, MCLCNC-RF, MCLCNC-SVM, all-RF, all-SVM
df.shape
```

```python
df_mcnc = df.iloc[:220,:]
df_mclcnc = df.iloc[220:440,:]
df_all = df.iloc[440:,:]
```

```python
#separate rf and svm, unscrambled
rf_indices = np.arange(0,20,2)
svm_indices = np.arange(1,20,2)
```

```python
bplot = plt.boxplot([df_mcnc.iloc[rf_indices,6],df_mcnc.iloc[svm_indices,6],
                     df_mclcnc.iloc[rf_indices,6],df_mclcnc.iloc[svm_indices,6],
                     df_all.iloc[rf_indices,6],df_all.iloc[svm_indices,6]],
            positions=[1,1.4,1.8,2.2,2.6,3], widths=0.2,patch_artist=True,medianprops=dict(color='black'))

plt.xlim(0.8,3.8)
plt.xticks([1.2,2,2.8],labels=['MCNC','MCLCNC','all'])

colors = ['lightgreen','pink','lightgreen','pink','lightgreen','pink','lightgreen','pink']
for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_edgecolor('black')

lightgreen_patch = mpatches.Patch(color='lightgreen', label='RF')
pink_patch = mpatches.Patch(color='pink', label='SVM')
plt.legend(handles=[lightgreen_patch,pink_patch])        
plt.title('ROC AUC')

plt.show()
```
