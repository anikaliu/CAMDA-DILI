```python
RF_test = {}
SVM_test = {}
i=0
for file in glob.glob('./**/ambiguous_results.csv'):
    model = file.split('/')[1]
    print(file)
    Test_data = pd.read_csv(file)
    #Test_data.drop(Test_data.columns[0],axis=1,inplace=True) # drop index column
    RF_models_ts = Test_data[Test_data['splits'].str.contains("MCNC.RF")]
    SVM_models_ts = Test_data[Test_data['splits'].str.contains("MCNC.SVM")]

    
    RF_test['RF_'+model+'_external_test'] = [str(round(RF_models_ts.mean(axis=0)[i],3)) +"\u00B1"+ str(round(RF_models_ts.std(axis=0)[i],3)) for i in range(len(RF_models_ts.mean(axis=0)))] 
    SVM_test['SVM_'+model+'_external_test'] = [str(round(SVM_models_ts.mean(axis=0)[i],3)) +"\u00B1"+ str(round(SVM_models_ts.std(axis=0)[i],3)) for i in range(len(SVM_models_ts.mean(axis=0)))] 
```
